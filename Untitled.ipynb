```python
mp  = 1836. # proton mass
V_s = 0.2  # in unit [c]
K   = 10.0 # electric field ratio
px  = np.linspace(-1,1,201)
xi  = np.linspace(-12,12,201)

xi, px = np.meshgrid(xi, px)
Hami= (px**2+1)**0.5*mp-V_s*px*mp+0.5*K*(xi-2)**2

levels = np.linspace(np.min(Hami), np.max(Hami), 41) 

plt.contourf(xi, px, Hami, levels=levels, cmap='jet')
cbar=plt.colorbar(pad=0.01)
plt.contour(xi, px, Hami, colors='k', style=':')

#### manifesting colorbar, changing label and axis properties ####
cbar.set_label('H',fontdict=font)        
plt.xlabel(r'$\xi$',fontdict=font)
plt.ylabel(r'$p_x$',fontdict=font)
#plt.xlim(-3.5,0.5)
#plt.ylim(0,200)
plt.xticks(fontsize=20); 
plt.yticks(fontsize=20);
fig = plt.gcf()
fig.set_size_inches(10, 8)
```

```python
def pxpy_to_energy(gamma, weight):
    binsize = 100
    en_grid = np.linspace(0,250,100)
    en_bin  = np.linspace(0,250,101)
    en_value = np.zeros_like(en_grid)
    for i in range(binsize):
        en_value[i] = sum(weight[ (en_bin[i]<=gamma) & (gamma<en_bin[i+1]) ])
    return (en_grid, en_value)

part_number=5000
nsteps=40
v_s   =0.2
insert1='./test_particle/run_002_revise/'
insert_n='_0000'
t1=np.loadtxt(insert1+'t'+insert_n+'.txt')
x1=np.loadtxt(insert1+'x'+insert_n+'.txt')
px1=np.loadtxt(insert1+'px'+insert_n+'.txt')
t=np.reshape(t1,(part_number,nsteps))/(2.0*np.pi)
x=np.reshape(x1,(part_number,nsteps))/(2.0*np.pi)
px=np.reshape(px1,(part_number,nsteps))

px=px[(-3<x[:,0]) & (x[:,0]<3),:]
x=x[(-3<x[:,0]) & (x[:,0]<3),:]
ww=np.zeros_like(x)+1.0
ek=((px**2+1.0)**0.5-1)*1836.*0.51

dist_x1, den1 = pxpy_to_energy(ek[:,-1],ww[:,-1])
plt.subplot(1,1,1)
plt.plot(dist_x1,den1,color='black',linewidth=3)
plt.xlabel(r'$\varepsilon_p$'+' [MeV]',fontdict=font)
plt.ylabel('dN/d'+r'$\varepsilon_p$'+' [A.U.]',fontdict=font)
plt.xticks(fontsize=font_size);
plt.yticks(fontsize=font_size);
plt.yscale('log')
plt.ylim(1e0,1.5e3)

plt.grid(which='major',linestyle=':', linewidth=0.2, color='k')


plt.subplots_adjust(left=0.18, bottom=0.24, right=0.98, top=0.96, wspace=0.05, hspace=0.03)

fig = plt.gcf()
fig.set_size_inches(8, 4)
plt.show()
fig.savefig('./theory_sheet_dN_de.png',format='png',dpi=160, transparent=True)
plt.close("all")
```
